# **Redes Neuronales Artificiales**

**1. Redes Neuronales Biológicas**

Se entiende por BNN (*Red Neuronal Biológica* por sus siglas en inglés) al modelo simplificado del funcionamiento de una neurona biológica y su arquitectura dentro de una red interconectada de neuronas que se toma como base para definir la arquitectura y funcionamiento de las Redes Neuronales Artificiales (*ANN* por sus siglas en inglés).

Desde un punto de vista biológico, la neurona es una célula que constituye la unidad fundamental del sistema nervioso. Se compone de un cuerpo y de terminaciones que la conectan con su entorno. Las terminaciones a través de las cuales una neurona recibe un estimulo se conocen como *dendritas*, y la terminación a través de la cual transmiten su propia señal se conoce como *axón*. Por lo general, cada axón se divide en su extremo final en varias terminaciones llamadas *telodendrias*. Las telodendrias de una neurona se conectan a las dendritas de otra neurona estableciendo una profunda conexión entre ambas que se conoce como *sinapsis*.

Cada neurona biológica recibe un pequeño impulso eléctrico como estimulo de otras neuronas a través de su conexión sináptica que llamaremos *señal*. Cuando una neurona recibe una señal genera luego una señal propia que transmite a través de su axón a otras neuronas con las que comparte conexión sináptica. Para que genere esta señal propia, el estímulo que recibe la neurona debe superar cierto valor límite que lleva a la neurona a un estado de activación, desde el cual produce su señal. *Cada  neurona tiene una activación particular, por lo que cada una responde a estimulos diferentes*.

Al unir cada neurona biológica a miles de otras neuronas en una basta red interconectada se forma la Red Neuronal Biológica, que constituye la arquitectura fundamental de funcionamiento del cerebro humano.

bnn[1]

> La *BNN* humana es capaz de realizar operaciones y procedimientos altamente complejos que determinan cada aspecto de la vida humana. Si bien su estudio es aún un campo activo de investigación, el mapeo de algunas partes del cerebro humano sugiere que las neuronas usualmente se organizan en capas consecutivas.

**2. Redes Neuronales Artificiales**

> La estructura de una Red Neuronal Artificial desarrollada en la computación proviene directamente de la estructura de la Red Neuronal Biológica que compone el cerebro humano.

Una neurona artificial recibe un conjunto específico de *señales de entrada* $(x_1, x_2, ..., x_n)$, cada una de las cuales corresponde a la *señal de salida* de otra neurona. Cada entrada $x_i$ se multiplica por el peso (*weight*) de la conexión correspondiente $w_i$ *que representa la fuerza de la conexión sináptica entre dos neuronas biológicas*. Las entradas pesadas se suman entre sí para formar *la señal total que recibe la neurona*. Cada neurona cuenta con un sesgo $b_0$ (*bias*), un valor límite que se añade a la suma anterior, *y que representa la sensibilidad de la neurona biológica a la señal recibida*, imponiendo un valor de referencia propio de cada neurona artificial. La sumatoria de las entradas pesadas y el sesgo representan la recepción de la señal en el cuerpo de una neurona biológica.

Finalmente, esta señal recibida se evalúa dentro de la *función de activación ($\sigma$)* de la neurona artificial para generar una señal de salida *y*:

$$
\sigma \left(\sum_{i=1}^{n} w_i x_i + b_0 \right) = y
$$

*La función de activación representa la señal generada por la neurona biológica frente al estimulo recibido*. Esta salida "*y*" de la neurona artificial se transforma luego en la señal de entrada de otra neurona artificial.


artificial_neuron[2]

Teniendo en cuenta que, *cada neurona biológica en una BNN esta conectada a otras muchas neuronas y que, al parecer, se organizan por capas*, podemos construir una Red Neuronal Artificial conectando entre sí las salidas y entradas de neuronas artificiales y organizándolas por capas. Dentro de las ANN se distinguen principalmente tres capas:

- **Capa de Entrada:** Es la primera capa, toma los datos que van a procesarse y los organiza en el formato adecuado para ser transmitidos a través de la red.
- **Capa de Salida**: Es la última capa, recibe los datos procesados a través de toda la red y ajusta su formato para producir una respuesta final basada en los patrones identificados.
- **Capas Ocultas**: Son todas las capas intermedias entre la capa de entrada y la capa de salida. Se encargan de procesar la información e identificar los patrones inherentes para la tarea que se pretende modelar con el algoritmo.

mlp[3]

Con base en esta estructura podemos identificar dos tipos de ANN principales:

1. Perceptrón (SLP): ANN más básica y fundamental. Corresponde a la unión directa entre la capa de entrada y la capa de salida.

2. Perceptrón Multi-Capa (MLP): Formalmente una Red Neuronal, compromete una o más capas ocultas entre la capa de entrada y la capa de salida. Primer algoritmo del Deep Learning.

**3. ¿Cómo se entrena una Red Neuronal Artificial?**

1. Se inicializan de forma aleatoria todos los parámetros (pesos y sesgos) de la Red Neuronal. Es muy importante que esta inicialización sea completamente aleatoria.

2. *Forward Propagation*: Se alimenta la red con los datos de entrenamiento. La señal de entrada se utiliza para calcular la salida de la primera capa, que puede expresarse matricialmente de la forma:

$$
y = f(XW^{T} + b)
$$

Donde $f()$ es la función de activación, $X$ es el vector de entrada, $W$ es el vector de pesos y $b$ es el vector de sesgos. El vector y de salida se toma como entrada de la segunda capa oculta, se calcula su salida que a su vez se toma como entrada de la siguiente capa y asi sucesivamente se procesa la señal de entrada hasta la capa de salida. Finalmente, la capa de salida retorna la predicción de la red.

3. *Cálculo del Error*: Una vez obtenida la predicción de la red, el algoritmo calcula el error en la predicción haciendo uso de la función de pérdida definida:

$$
 Error = L(\hat{y}, y)
$$

Bajo esta definición, el algoritmo calcula la contribución de cada neurona en la capa de salida al error en la predicción a través del gradiente de la función de pérdida respecto a los parámetros:

$$
\nabla Error = \left( \frac{\partial L}{\partial f^{(k)}} \frac{\partial f^{(k)}}{\partial z^{(k)}} \frac{\partial z^{(k)}}{\partial W^{(k)}} , \frac{\partial L}{\partial f^{(k)}} \frac{\partial f^{(k)}}{\partial z^{(k)}} \frac{\partial z^{(k)}}{\partial b^{(k)}} \right)
$$

4. *Backpropagation*: Luego de calcular la contribución al error que proporciona la capa de salida, el algoritmo continua calculando el el aporte al error de cada capa anterior hasta alcanzar la capa de entrada, es decir, calcula el gradiente del error respecto a los pesos y sesgos de cada capa de la red propagando el error desde la capa de salida hasta la capa de entrada, **propagando el error hacia atrás**. De esta forma se obtiene el aporte global de la red al error en la predicción. Por ejemplo, para la capa inmediatamente anterior a la capa de salida el gradiente correspondiente es de la forma:

$$
\nabla Error = \left( \frac{\partial L}{\partial f^{(k)}} \frac{\partial f^{(k)}}{\partial z^{(k)}} \frac{\partial z^{(k)}}{\partial f^{(k-1)}}  \frac{\partial f^{(k-1)}}{\partial z^{(k-1)}}  \frac{\partial z^{(k-1)}}{\partial W^{(k-1)}}, \frac{\partial L}{\partial f^{(k)}} \frac{\partial f^{(k)}}{\partial z^{(k)}} \frac{\partial z^{(k)}}{\partial f^{(k-1)}} \frac{\partial f^{(k-1)}}{\partial z^{(k-1)}}  \frac{\partial z^{(k-1)}}{\partial b^{(k-1)}} \right)
$$

5. *Actualización de los parámetros*: Una vez calculado el gradiente del error respecto a los parámetros de cada capa, el algoritmo ejecuta un paso de Descenso de Gradiente para ajustar cada peso y sesgo de todas las neuronas de la red:

$$
W_{i}^{(l)} = W_{i}^{(l)} - \alpha \left(\frac{\partial Error}{\partial W^{(l)}} \right) \\
b_{i}^{(l)} = b_{i}^{(l)} - \alpha \left(\frac{\partial Error}{\partial b^{(l)}} \right) 
$$

6. Este proceso se repite iterativamente hasta obtener un error ajustado a la tolerancia definida o hasta que se alcance el número de iteraciones máximo definido.

Para explorar el notable potencial de las Redes Neuronales Artificiales (ANN), en este laboratorio entrenaremos un Perceptrón Multicapa (MLP) para abordar una tarea de Clasificación utilizando los mismos datos. Para ello, utilizaremos la clase *MLP_Classifier* de la librería *Scikit-Learn*, que permite implementar de manera sencilla una Red Neuronal Artificial capaz de resolver problemas en los que la Regresión Logística no resulta efectiva. 

En primer lugar importamos el modelo y los datos de entrenamiento:

```python
from sklearn.neural_network import MLPClassifier
import pandas as pd
import matplotlib.pyplot as plt

# Importamos los datos de entrenamiento que corresponden a la distribución que no es separable linealmente

data_df = pd.read_csv("classification_train.txt", sep='\t')
data_predictors = data_df[["X","Y"]] # Generamos una matriz con los datos de los predictores X,Y
data_label = data_df["Z"].map(lambda x: 1 if x > 0 else 0) # Vector con las etiquetas. En este caso no tomamos -1, sino 0
```

A continuación, definimos un modelo de MLP con una capa oculta compuesta por 4 neuronas y una función de activación ReLu:

```python
# Establecemos adicionalmente un límite máximo de 10000 iteraciones para el proceso de entrenamiento
modelo_classification = MLPClassifier(hidden_layer_sizes=(4,),activation="relu",max_iter=10000)
```

Entrenamos el modelo:

```python
""" 
El primer argumento corresponde a la matriz X, en las que cada columna se corresponde con el predictor
y cada fila con cada instancia. Tal y como la construimos en el laboratorio anterior.

Por su parte, el segundo argumento corresponde con el vector que contiene las etiquetas.

Importante: Ambos deben ser arreglos de Numpy.
"""
modelo_classification.fit(data_predictors, data_label)
```

Finalmente, utilizamos el modelo entrenado para obtenerpredicciones sobre los datos de prueba y graficamos los resultados:

```python
# Importamos los datos de prueba
data_df = pd.read_csv("classification_test.txt", sep='\t')
data_predictors_test = data_df[["X","Y"]] # Generamos una matriz con los datos de los predictores X,Y
data_label_test = data_df["Z"].map(lambda x: 1 if x > 0 else 0) 

# Utilizamos el modelo entrenado sobre los datos de prueba
predictions = modelo_classification.predict(data_predictors_test)

# Graficamos los resultados: A la derecha los datos reales y a la izquierda las predicciones
fig,ax = plt.subplots(1,2,figsize=(10,5),layout="constrained")
data_predictors_test.plot.scatter("X","Y",c=data_label_test,ax=ax[0],title="Datos originales",cmap="coolwarm",colorbar=False)
data_predictors_test.plot.scatter("X","Y",c=predictions,ax=ax[1],title="Predicción MLP",cmap="coolwarm",colorbar=False)
plt.show()
```

En el ejercicio anterior usamos una función de activación *ReLu* que ejecuta una transformación no lineal a los datos y actúa de forma similar a la función logística, base de la Regresión Logística. Veamos que sucede si utilizamos una función de activación lineal en su lugar:

```python
# Modificamos la función de activación con --identity-- para introducir linealidad
modelo_classification_2 = MLPClassifier(hidden_layer_sizes=(4,),activation="identity",max_iter=10000)
modelo_classification_2.fit(data_predictors,data_label)
```

```python
# Utilizamos el modelo entrenado nuevamente sobre los datos de prueba
predictions_2 = modelo_classification_2.predict(data_predictors_test)

# Graficamos los resultados: A la derecha los datos reales y a la izquierda las predicciones
fig,ax = plt.subplots(1,2,figsize=(10,5),layout="constrained")
data_predictors_test.plot.scatter("X","Y",c=data_label_test,ax=ax[0],title="Datos originales",cmap="coolwarm",colorbar=False)
data_predictors_test.plot.scatter("X","Y",c=predictions_2,ax=ax[1],title="Predicción MLP",cmap="coolwarm",colorbar=False)
plt.show()
```

Observe que en este caso el MLP impone una separación lineal en el espacio. Identificó cada clase y las separo linealmente, pese a que su distribución inicial era polar !!

**5.** Observe que los resultados varían para la función de activación ReLu pero no para la función de activación lineal. ¿A qué se debe esto?

Cuando usamos ReLU como función de activación, la red neuronal puede aprender relaciones no lineales en los datos. Esto es importante porque nuestro conjunto de datos no es linealmente separable. Al agregar más neuronas ocultas, la red con ReLU mejora su capacidad para ajustarse a patrones más complejos y separar mejor las clases.

Por otro lado, cuando usamos una función de activación lineal (identity), aunque agreguemos más neuronas, todas las capas aplican transformaciones lineales. Como resultado, la red sigue siendo lineal, y no puede aprender relaciones no lineales. Por eso, su rendimiento no mejora mucho al aumentar las neuronas.

# **MLP para Clasificación de Imágenes (opcional)**

El conjunto de datos *Optical Recognition of Handwritten Digits* esta compuesto por imágenes de los dígitos de 0 a 9 escritos a mano. En esta sección opcional del laboratorio vamos a entrenar un modelo de red neuronal que nos permita clasificar el dígito mostrado en la imagen. 
Para mayor información sobre este conjunto de datos puede consultar el siguiente [enlace](https://archive.ics.uci.edu/dataset/80/optical+recognition+of+handwritten+digits).

En primer lugar, importamos el dataset:

```python
# Importamos el conjunto de datos desde Scikit-Learn
from sklearn.datasets import load_digits

data_digits = load_digits()
# La llave -data- contiene las imágenes de cada dígito de la base de datos
digits = data_digits["data"]
# La llave -target- contiene las etiquetas de clasificación de cada dígito
labels = data_digits["target"]
```

```python
""" Información adicional de la base de datos"""
print(data_digits['DESCR'])
```

Ahora observemos algunas imágenes:

```python
fig,ax = plt.subplots(1,4,figsize=(20,5))
for i in range(4):
  ax[i].imshow(digits[6*i].reshape((8,8)),cmap="gray")
  ax[i].set_title(f"Digito: {labels[6*i]}")
  ax[i].axis("off")
plt.show()
```

Para entrenar el modelo vamos a separar el conjunto de datos en un conjunto de entrenamiento (80%) y un conjunto de prueba (20% del dataset). Para ello tulizamos la función *train_test_split()*:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# En test_size específicamos el tamaño (como porcentaje) del conjunto de prueba
# La varible random_state es muy importante, Asegura que el proceso aleatorio con el que se construyen
# los dos conjuntos sea "almacenado" de modo que cada vez que se ejecute el proceso los conjuntos tengan 
# los mismos elementos
X_train, X_test, y_train, y_test = train_test_split(digits,labels,test_size=0.2, random_state=42)
```

Como modelo utilizamos un MLP para Clasificación con 2 neuronas en la capa oculta y una función de activación *ReLu*:

```python
modelo_class = MLPClassifier((2,),max_iter=10000,activation="relu")
# Entrenamos el modelo con los datos de entrenamiento
modelo_class.fit(X_train,y_train)
```

Por último, vamos a evaluar el desempeño del modelo.

```python
# Hallamos las predicciónes sobre el conjunto de prueba
predicciones = modelo_class.predict(X_test)
print("Precisión:", accuracy_score(y_test,predicciones))
```

Así, mediante el uso de un MLP logramos ajustar un modelo capaz de clasificar dígitos escritos a mano con una precisión del 28%.
